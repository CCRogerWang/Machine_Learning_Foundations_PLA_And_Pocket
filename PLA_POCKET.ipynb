```python
import pandas as pd
import io
import requests
import numpy as np
from random import randrange, uniform
from random import randint
import random
```

```python
def fetchData(url):
    s = requests.get(url).content
    return pd.read_csv(io.StringIO(s.decode('utf-8')), sep="\s+", names=['c1','c2','c3','c4','l1'])
```

```python
url_training = "https://www.csie.ntu.edu.tw/~htlin/mooc/datasets/mlfound_math/hw1_18_train.dat"
url_testting = "https://www.csie.ntu.edu.tw/~htlin/mooc/datasets/mlfound_math/hw1_18_test.dat"
df_training = fetchData(url_training)
df_testing = fetchData(url_testting)
```

```python
def createXandYData(dataFrame):
    X = dataFrame[['c1','c2','c3','c4']] 
#     X.insert(0, "c0", 1)
    X['w0'] = 1.0 #  很重要！
    Y = dataFrame[['l1']]
    return X, Y
```

```python
def numberOfErrors(X, Y, w):
    errors = 0
    for i in range(len(X)):
        x = X.loc[i].values
        y = Y.loc[i].values
        if isError_SIGN(w, x, y):
            errors += 1
    return errors
```

```python
def isError_SIGN(w, x, y):
    return (np.sign(np.dot(w, x)) != y)
```

```python
def evl_error_rate(X, Y, w):
    data_size = len(X)
    errors = 0
    for i in range(data_size):
        x = X.loc[i].values
        y = Y.loc[i].values
        if (np.sign(np.dot(w, x)) != y):
            errors = errors + 1
    return errors/data_size
```

```python
def PLA_POCKET(X, Y, iters):
    w = np.zeros(X.shape[1])
    best_error_rate = evl_error_rate(X, Y, w)
    current_iters = 0
    w_pocket = 0
    while current_iters <= iters:
        i = X.sample(1).index
        x = X.loc[i].values[0] #取 row data
        y = Y.loc[i].values[0] #取 label
        if (np.sign(np.dot(w, x)) != y): #有錯
            current_iters = current_iters + 1
            w = w + 1*y*x # 就修正
            error_rate_new = evl_error_rate(X, Y, w)
            if error_rate_new < best_error_rate: #比較好就放口袋
                w_pocket = w
                best_error_rate = error_rate_new
    return w_pocket

#找到錯誤了 每次的w 都要更新 如果比之前的好就放入口袋 才是正確的邏輯
```
