```python
def createXandYData(dataFrame):
    X = dataFrame[['c1','c2','c3','c4']] 
    X.insert(0, "c0", 1)
#     X['w0'] = 1 #  很重要！
    Y = dataFrame[['l1']]
    return X, Y
```

```python
#是否有錯誤產生
def isError(w, x, y):
    return True if np.dot(w,x)*y <= 0 else False
```

```python
#更正錯誤
def updateW(w_t, y_t, x_t):
    w_t += 0.5*y_t*x_t
    return w_t
```

```python
#PLA 學習
def PLA(X, Y):
    w = np.zeros(X.shape[1])
    step = 0
    while True:
        errors = 0
        for i in range(len(X)):
            x = X.loc[i].values
            y = Y.loc[i].values
            if isError(w, x, y):
                step += 1
                errors += 1
                w = updateW(w, y, x)
        if errors == 0:
            break
    return step, w
```

```python
X, Y = createXandYData(df)
PLA(X, Y)
```

```python
step = 0
sizeOfRange = 2000
for i in range(sizeOfRange):
    df = df.sample(frac=1).reset_index(drop=True)
    df.head()
    X, Y =createXandYData(df)
    tmpstep, w = PLA(X, Y)
    step += tmpstep
    print('step '+ str(i)+ ' ' + str(tmpstep))
print('mean ' + str(step/sizeOfRange))
```

```python
def isError_SIGN(w, x, y):
    return (np.sign(np.dot(w, x)) != y)
```

```python
def PLA_POCKET(X, Y, iters):
    w_best = np.zeros(X.shape[1])
    best_error_rate = evl_error_rate(X, Y, w_best)
    current_iters = 0
    while current_iters <= iters:
        i = X.sample(1).index
        x = X.loc[i].values[0] #取 row data
        y = Y.loc[i].values[0] #取 label
        if (np.sign(np.dot(w_best, x)) != y):
            w_new = w_best + 1*y*x
            error_rate_new = evl_error_rate(X, Y, w_new)
            print(error_rate_new)
            if error_rate_new < best_error_rate:
                w_best = w_new
                best_error_rate = error_rate_new
                print(str(current_iters) + " " + str(w_best) + " " + str(best_error_rate))
        current_iters = current_iters + 1
    return w_best
```

```python
def evl_error_rate(X, Y, w):
    data_size = len(X)
    errors = 0
    for i in range(data_size):
        x = X.loc[i].values
        y = Y.loc[i].values
        if isError(w, x, y):
            errors += 1
    return errors/data_size
```

```python
def pocket_pla(datas, limit):  
    ###############
    def _calc_false(vec):
        res = 0
        for data in datas:
            t = np.dot(vec, data[0])
            if np.sign(data[1]) != np.sign(t):
                res += 1
        return res
    ###############
    w = [0 ,0 ,0 ,0 ,0]
    least_false = _calc_false(w)
    res = w

    for i in range(limit):
        data = random.choice(datas)
        t = np.dot(w, data[0])
        if np.sign(data[1]) != np.sign(t):
            t = w + data[1] * data[0]
            t_false = _calc_false(t)

            w = t

            if t_false <= least_false:
                least_false = t_false
                res = t
    return res, least_false
```
